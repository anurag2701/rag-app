```python
import ollama
import os

messages = [
    {'role': 'user', 'content': 'Hello!'},
    {'role': 'assistant', 'content': 'Hi there! How can I help you today?'},
    {'role': 'user', 'content': 'What is the sky blue?'},
]

ollama.api_url = os.getenv("OLLAMA_API_URL", "http://localhost:11434")
ollama.chat(model='llama3', messages=messages)


# response = ollama.chat(model='gemma3', messages=messages)
# print(response.message.content)
```
